```bash
cellranger mkfastq  --id=tiny-bcl2 \
                    --run=../data/cellranger-tiny-bcl-1.2.0 \
                    --samplesheet=../data/cellranger-tiny-bcl-samplesheet-1.2.0.csv \
#                     --localcores=$SLURM_CPUS_PER_TASK \
#                     --localmem=34
```
