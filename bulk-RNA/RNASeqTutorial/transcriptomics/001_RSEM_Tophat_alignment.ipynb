# there are different ways you can generate index file for bam, here is using picard

```bash
module load picard
```

```bash
java -Xmx4g -XX:ParallelGCThreads=5 -jar $PICARDJARPATH/picard.jar BuildBamIndex I=tophat/th.quant.ctrl1.bam O=tophat/th.quant.ctrl1.bam.bai
java -Xmx4g -XX:ParallelGCThreads=5 -jar $PICARDJARPATH/picard.jar BuildBamIndex I=tophat/th.quant.ctrl2.bam O=tophat/th.quant.ctrl2.bam.bai
java -Xmx4g -XX:ParallelGCThreads=5 -jar $PICARDJARPATH/picard.jar BuildBamIndex I=tophat/th.quant.ctrl3.bam O=tophat/th.quant.ctrl3.bam.bai
java -Xmx4g -XX:ParallelGCThreads=5 -jar $PICARDJARPATH/picard.jar BuildBamIndex I=tophat/th.quant.expr1.bam O=tophat/th.quant.expr1.bam.bai
java -Xmx4g -XX:ParallelGCThreads=5 -jar $PICARDJARPATH/picard.jar BuildBamIndex I=tophat/th.quant.expr2.bam O=tophat/th.quant.expr2.bam.bai
java -Xmx4g -XX:ParallelGCThreads=5 -jar $PICARDJARPATH/picard.jar BuildBamIndex I=tophat/th.quant.expr3.bam O=tophat/th.quant.expr3.bam.bai
```

**procedures for IGV
1. on local computer, install IGV
2. open IGV
3. select genome, (for customized geneome, give a name, and give the fastq file)
4. drag and drop (sorted and indexed) .bam files and .tdf files to the window. 
5. search for Fgf21 to visualize the alignment, for both RSEM and Tophat2 alignment.
